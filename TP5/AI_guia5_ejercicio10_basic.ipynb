```
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```
data = np.loadtxt("seriesautos.txt", dtype=np.float32)
X_train, Y_train = data[0:83].reshape(1, -1, 1), data[83].reshape(1, -1, 1)
```

```
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('loss')<0.0005):
      print("\nYa se alcanzó una 'loss' igual a 0.0005 entonces paramos!")
      self.model.stop_training = True

callbacks = myCallback()
```

```
model = keras.models.Sequential([
    keras.layers.LSTM(20, return_sequences=True, input_shape=[None, 1]),
    keras.layers.BatchNormalization(),
    keras.layers.LSTM(20),
    keras.layers.BatchNormalization(),
    keras.layers.Dense(1)
])

model.compile(loss="mse", optimizer="adam")
history = model.fit(X_train, Y_train, epochs=2500, verbose=0, callbacks=[callbacks])
```

```
X = X_train
for paso in range(10):
  y_pred_one = model.predict(X)[:, np.newaxis, :]
  X = np.concatenate([X, y_pred_one], axis=1)
```
