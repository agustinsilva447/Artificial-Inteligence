```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal
```

image.png

```python
N = 100
dy = 1/N
y = np.linspace(0, 1, N+1)
```

```python
def tren_pulsos(p,e):
  V = np.zeros(len(y))
  c = 0
  for i in range(len(y)):
    if c < p:
      V[i] = 0
      c += 1
    elif c < 2*p:
      V[i] = e
      c += 1
    else:
      c = 0
      V[i] = e
  return V

def mL2V(y):
  p = 34    # periodo sobre 2 del tren
  e = 100   # valor max del potencial
  return tren_pulsos(p,e)

  #return 1000 * (y - 0.5)**2
  #return 1000 * np.exp(-(y-0.7)**2/(2*0.05**2))
```

```python
plt.figure(figsize=(10,5))
plt.plot(y, mL2V(y))
plt.title('Potential', fontsize=30)
plt.ylabel('$V/V_0$',  fontsize=20)
plt.xlabel('$x/L$',    fontsize=20)
plt.grid()
plt.plot()

plt.savefig('v3p1.png', dpi=200)
```

```python
d =  (1)/(dy**2)   + mL2V(y)[1:-1]
e = -(1)/(2*dy**2) * np.ones(len(d)-1)
```

```python
w, v = eigh_tridiagonal(d, e)
```

```python
plt.figure(figsize=(10,5))
for i in range(1):
  plt.plot(y[1:-1], v.T[i]**2, label = "Solution {}".format(i))

plt.title('Eigenstates', fontsize=30)
plt.ylabel('$\psi$',     fontsize=20)
plt.xlabel('$x/L$',      fontsize=20)
plt.grid()
plt.legend()
plt.plot()

plt.savefig('v3p2.png', dpi=200)
```
